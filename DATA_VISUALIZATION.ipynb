```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_excel('/content/merged_data_temp_co2_precipitate.xlsx')
```

```python
data.head()
```

```python
Countries_list = data["Country"].unique()
```

```python
# top 5 country with highest co2 record

top_5_countries = data['Country'].value_counts().nlargest(5).index

for country in top_5_countries:
    country_data = data[data['Country'] == country]
    plt.figure(figsize=(10, 6))

    plt.plot(country_data['Year'], country_data['Annual CO₂ emissions'])
    plt.xlabel('Year')
    plt.ylabel('CO2 emissions')
    plt.title(f'CO2 emissions over time in {country}')
    plt.show()
```

```python
# top 5 country with highest Precipitate record

top_5_countries = data['Country'].value_counts().nlargest(5).index

for country in top_5_countries:
    country_data = data[data['Country'] == country]
    plt.figure(figsize=(10, 6))

    plt.plot(country_data['Year'], country_data['Precipitate'])
    plt.xlabel('Year')
    plt.ylabel('Precipitate')
    plt.title(f'Precipitate over time in {country}')
    plt.show()
```

```python
# top 5 country with highest Temperature record

top_5_countries = data['Country'].value_counts().nlargest(5).index

for country in top_5_countries:
    country_data = data[data['Country'] == country]
    plt.figure(figsize=(10, 6))

    plt.plot(country_data['Year'], country_data['Temperature'])
    plt.xlabel('Year')
    plt.ylabel('Temperature')
    plt.title(f'Temperature	over time in {country}')
    plt.show()
```

```python
def plot_with_separate_scales(country, data):
    country_data = data[data['Country'] == country]
    fig, ax1 = plt.subplots(figsize=(10, 6))

    color = 'tab:red'
    ax1.set_xlabel('Year')
    ax1.set_ylabel('CO2 emissions', color=color)
    ax1.plot(country_data['Year'], country_data['Annual CO₂ emissions'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()
    color = 'tab:blue'
    ax2.set_ylabel('Temperature', color=color)
    ax2.plot(country_data['Year'], country_data['Temperature'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    ax3 = ax1.twinx()
    ax3.spines['right'].set_position(('outward', 60))
    color = 'tab:green'
    ax3.set_ylabel('Precipitate', color=color)
    ax3.plot(country_data['Year'], country_data['Precipitate'], color=color)
    ax3.tick_params(axis='y', labelcolor=color)

    plt.title(f'CO2 Emissions, Temperature, and Precipitation over time in {country}')
    fig.tight_layout()
    plt.show()

# top 5 country with highest co2 record
top_5_countries = data['Country'].value_counts().nlargest(5).index

for country in top_5_countries:
    plot_with_separate_scales(country, data)
```
